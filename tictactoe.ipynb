```python
import random

#create a frame for starting empty board
board = [' ', ' ', ' ',
        ' ', ' ', ' ',
        ' ', ' ', ' ']

#we create a function to print board to help players keep tabs
def print_board(board):
    print(board[0] + ' | ' + board[1] + ' | ' + board[2])
    print('---------')
    print(board[3] + ' | ' + board[4] + ' | ' + board[5])
    print('---------')
    print(board[6] + ' | ' + board[7] + ' | ' + board[8])
    print('---------')

#some global variables
player_1 = 'X'
player_2 = 'O'
current_player = 'X'
winner = None
game_running = True

#I have made 2 variations of the game; 2 human players or 1 player playing against computer
def players_input(board): #this is 2 player game
    input_1 = int(input("Player 1, please choose number from 1 to 9: "))
    if input_1 >= 1 and input_1 <=9 and board[input_1 - 1] == ' ': #this if conidition ensures numbers choosen are between 1 and 9 and the position choosen is empty
        board[input_1 - 1] = player_1
    else:
        print("Please choose number from 1 to 9 and make sure spot isn't already picked :) ")
    input_2 = int(input("Player 2, please choose number from 1 to 9: "))
    if input_2 >= 1 and input_2 <=9 and board[input_2 - 1] == ' ': #Input for player 2
        board[input_2 - 1] = player_2
    else:
        print("Please choose number from 1 to 9 and make sure spot isn't already picked :) ")

def player_input(board): #for one player playing against computer
    input_1 = int(input("Player 1, please choose number from 1 to 9: "))
    if input_1 >= 1 and input_1 <=9 and board[input_1 - 1] == ' ':
        board[input_1 - 1] = player_1
    else:
        print("Please choose number from 1 to 9 and make sure spot isn't already picked :) ")

def switch_player(board): #This function will come in handy when alternating between the player and computer
    global current_player
    if current_player == 'X':
        current_player = 'O'
    else:
        current_player = 'X'


def computer(board): #help the computer input their choices uses random function
    while current_player == 'O':
        position = random.randint(0,8)
        if board[position] == ' ':
            board[position] = 'O'
            switch_player(board)
    

#create functions to check winner
def horizontal_check(board):
    global winner
    if board[0] == board[1] == board[2] and board[0] != ' ':
        winner = board[0]
        return True
    elif board[3] == board[4] == board[5] and board[3] != ' ':
        winner = board[3]
        return True
    elif board[6] == board[7] == board[8] and board[8] != ' ':
        winner = board[6]
        return True
def row_check(board):
    global winner
    if board[0] == board[3] == board[6] and board[0] != ' ':
        winner = board[3]
        return True
    elif board[1] == board[4] == board[7] and board[1] != ' ':
        winner = board[4]
        return True
    elif board[2] == board[5] == board[8] and board[2] != ' ':
        winner = board[2]
        return True

def diag_check(board):
    global winner
    if board[0] == board[4] == board[8] and board[0] != ' ':
        winner = board[8]
        return True
    elif board[2] == board[4] == board[6] and board[2] != ' ':
        winner = board[6]
        return True

def check_win(): #then combine them into one function that checks all possible win scenarios
    global game_running
    if diag_check(board) or horizontal_check(board) or row_check(board):
        #print_board(board)
        print(f'{winner} wins')
        game_running = False #we will run game on loop so changing this to false is crucial to end loop and consequently game once winner is determined


def check_tie(board): #to check whether game ends in a tie
    global game_running
    if ' ' not in board:
        print_board(board)
        game_running = False


print("Hello, please choose if you are playing alone of there are 2 players") #we need user to choose if there is 1 or 2 players
players = int(input("Choose number of players 1 or 2: "))

if players == 1: #for one player and computer
    while game_running:
        print_board(board)
        player_input(board)
        check_win()
        check_tie(board)
        switch_player(board)
        computer(board)
        check_win()
        check_tie(board)
elif players == 2: #for two human players
     while game_running:
        print_board(board)
        players_input(board)
        check_win()
        check_tie(board)   

    
```
